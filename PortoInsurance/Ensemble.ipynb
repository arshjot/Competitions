```python
import pandas as pd
import numpy as np
```

```python
v11 = pd.read_csv('./submit/v11_xgb_one_hot_standard_strat_lambda_100_val_2842.csv.gz')
v12 = pd.read_csv('./submit/v12_xgb_one_hot_standard_strat_lambda_100_poisson_val_2848.csv.gz')
```

```python
concat = pd.concat([v11, v12], axis=1)
```

```python
ensemble = v11
ensemble['target'] = concat[['target']].mean(axis=1)
```

```python
ensemble.to_csv('./submit/v14_ensemble_of_v11_v12.csv.gz', index=False, compression='gzip')
```
